```python
#Python program to list number of items in a dictionary key and sort the list with the help of  the function
dict1= {"a": 1, "b": 0}
sort_dict= dict(sorted(dict1.items(), key=lambda item: item[1])) 
print("Sorted Dictionary : ", sort_dict)
```

```python
#Python program to list number of items in a dictionary key and sort the list without a function
import operator
from operator import itemgetter
dict2 = { 'a':3,'b':2,'d':1,'c':0 }
sort_dict= dict(sorted(dict2.items(), key=operator.itemgetter(1))) 
print("Sorted Dictionary : ", sort_dict)
```

```python
#a Python program to find the repeated items of a list
lst = [10,8,4,5,2,1,3,6,8, 9,10,5,3,5 ,6,7]
dupItems = []
uniqItems = {}
for x in lst:
   if x not in uniqItems:
      uniqItems[x] = 1
   else:
      if uniqItems[x] == 1:
         dupItems.append(x)
      uniqItems[x] += 1
print("Duplicate items:",dupItems)
```
